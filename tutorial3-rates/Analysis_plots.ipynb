```python
import msm_we
import numpy as np
import pickle
```

```python
# Open the pickle file
with open('Outputs/SteadyState_Flux.obj', 'rb') as infile:
    msm_we_model = pickle.load(infile)
```

```python
cluster_centers = msm_we_model.targetRMSD_centers #Average progress cordinate value of all structures in the given clusters  
cluster_pSS = msm_we_model.pSS #Steady-State probaility Equation with transition matrix
rate_constant = msm_we_model.JtargetSS #Steady-State target flux/tau

print(f"The rate constant for folding is {rate_constant}/s")
```

```python
import matplotlib.pyplot as plt

plt.scatter(cluster_centers,cluster_pSS, color='k')

target_pcoord = cluster_centers[-1]
basis_pcoord = cluster_centers[-2]
target_pSS = cluster_pSS[-1]
basis_pSS = cluster_pSS[-2]
plt.scatter(target_pcoord, target_pSS, color='purple') #Target State prob
plt.scatter(basis_pcoord, basis_pSS, color='blue') #basis State prob

plt.title('NESS probability')
plt.yscale('log')
plt.ylabel('Probability')
plt.xlabel('MSM State Average Progress Coordinate')
```
